##### The GROUP BY function/clause is used to organize rows that have the same values in one or more specified columns into a single summary row. It is almost always used with aggregate functions (like COUNT(), SUM(), AVG(), MAX(), and MIN()) to perform calculations on each set of grouped data. 

```python
df = pd.read_csv(r"C:\Users\Riya Thokal\Downloads\Car_sales.csv")
```

```python
df
```

```python
df = df[['Manufacturer', 'Model' , 'Sales_in_thousands', 'Vehicle_type', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Fuel_capacity','Curb_weight']]
```

# Stacked Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Group by both categories and sum the sales
# Assumes column names: 'Outlet Location Type', 'Item Fat Content', 'Sales'
stacked_data = df.groupby(['Vehicle_type', 'Price_in_thousands'])['Sales_in_thousands'].sum()

# 2. Unstack the data to pivot 'Item Fat Content' into columns
stacked_data = stacked_data.unstack('Price_in_thousands')

# 3. Plot the stacked bar chart
stacked_data.plot(kind='bar', stacked=True, figsize=(10,6))

# 4. Customize the plot
plt.title('Outlet Type and Outlet Size')
plt.ylabel('Total Sales')
plt.xlabel('OutletType')
plt.xticks(rotation=90)
plt.legend(title='Outlet Type') # Optional: Adds title to legend
plt.tight_layout() # Ensures labels don't get cut off
plt.show()
```

# Bar Chart

```python
Price_in_thousands = df.groupby('Vehicle_type')['Sales_in_thousands'].sum().sort_values(ascending=False)

# Set figure size
plt.figure(figsize=(10,4))

# Create bar chart
bar = plt.bar(Price_in_thousands.index, Price_in_thousands.values, color='blue')

# Rotate x-axis labels for readability
plt.xticks(rotation=90)
plt.xlabel('Sales in thousands')
plt.ylabel('Vehicle_type')
plt.title('Price in thousands')
# Add value labels on top of each bar
for bar in bar:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center',
        va='bottom',
        fontsize=8
    )

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```
